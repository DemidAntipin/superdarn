# Создание 1-ой модели

```python
import tensorflow as tf
from tensorflow.keras import layers
from glob import glob
import os
```

```python
input_layer=layers.Input(shape=(100, 2), name="input")
lstm=layers.LSTM(64, return_sequences=True, name="lstm_encoder_1")(input_layer)
lstm=layers.LSTM(16, name="lstm_encoder_2")(lstm)
encoder_output=layers.Dense(3, activation="sigmoid")(lstm)
repeat=layers.RepeatVector(100)(encoder_output)
lstm=layers.LSTM(128, return_sequences=True, name="lstm_decoder")(repeat)
abs_output=layers.TimeDistributed(layers.Dense(1, activation="sigmoid"), name="abs_output")(lstm)
phase_output=layers.TimeDistributed(layers.Dense(1, activation="sigmoid"), name="phase_output")(lstm)
```

## Создание кастомный функций потерь, учитывающих качество точек

```python
def abs_qty(y_true, y_pred):
  qty=y_true[:, :,1]
  mae=tf.abs(y_true[:,:, 0]-tf.squeeze(y_pred))
  loss=mae*qty
  return tf.reduce_mean(loss)

def phase_qty(y_true, y_pred):
  qty = y_true[:,:,1]
  mae = tf.abs(y_true[:,:,0] - tf.squeeze(y_pred))
  loss = mae * qty
  return tf.reduce_mean(loss)
```

```python
autoencoder=tf.keras.models.Model(inputs=input_layer, outputs=[abs_output, phase_output])
autoencoder.summary()
autoencoder.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=1e-3), loss=[abs_qty,phase_qty], metrics=[abs_qty, phase_qty])
tf.keras.utils.plot_model(autoencoder, show_shapes=True, show_layer_names=True)
```

## Подготовка датасета

Class `DataGenerator` извлекает значения из полученного фрагмента данных, нормализует их к значениям от 0 до 1 и формирует элементы датасета.

```python
class DataGenerator:
    def __init__(self, chunk):
        self.chunk = chunk
    def __iter__(self):
        abs = []
        phases = []
        qty = []
        for i in range(3, self.chunk.shape[1], 3):
          abs.append(tf.convert_to_tensor(np.log(self.chunk.iloc[:, i].values+1)/10, dtype=tf.float32))
          phases.append(tf.convert_to_tensor((self.chunk.iloc[:, i + 1].values+np.pi)/2/np.pi, dtype=tf.float32))
          qty.append(tf.convert_to_tensor(self.chunk.iloc[:, i + 2].values, dtype=tf.float32))
        abs_tensor = tf.stack(abs, axis=1)
        phases_tensor = tf.stack(phases, axis=1)
        qty_tensor = tf.stack(qty, axis=1)
        X = tf.stack([abs_tensor, phases_tensor], axis=-1)
        Y_abs = tf.stack([abs_tensor, qty_tensor], axis=-1)
        Y_phases = tf.stack([phases_tensor, qty_tensor], axis=-1)
        for x, abs, phase in zip(X, Y_abs, Y_phases):
            yield (x, (abs, phase))
```

Фильтрация датасета удаляет элементы с бесконечными значениями в тензорах.

```python
def filter_dataset(x, y):
    return tf.reduce_all(tf.math.is_finite(x)) & tf.reduce_all(tf.math.is_finite(y[0])) & tf.reduce_all(tf.math.is_finite(y[1]))
```

Функция занимается созданием обучающего датасета. Генерация элементов происходит в классе `DataGenerator`, определенным выше, также как и фильтрация полученных элементов.

Функция определяет shape тензоров датасета:
+ Входной тензор включает значения амплитуды и фазы точки.
+ Выходной тензор состоит 2 других такой же формы. 1-ый содержит метки амплитуды точки и её качества. Качество точки добавлено в тензор истинных значений чтобы получить к нему доступ в функции потерь. 2-ой тензор содержит значения фазы вместо амплитуды и также качество точек.

```python
def create_tf_dataset(chunk):
    data_gen = DataGenerator(chunk)
    dataset = tf.data.Dataset.from_generator(
        lambda: data_gen,
        output_signature=(
            tf.TensorSpec(shape=(100, 2), dtype=tf.float32),
           (tf.TensorSpec(shape=(100, 2), dtype=tf.float32),
            tf.TensorSpec(shape=(100, 2), dtype=tf.float32)))
       )
    dataset=dataset.filter(filter_dataset)
    return dataset
```

## Сборка модели

```python
input_layer = layers.Input(shape=(100, 2), name="Input_Signal")
abs_0 = input_layer[:, 0, 0]
abs_0 = tf.keras.layers.Reshape((1,1))(abs_0)
lstm_encoder = layers.LSTM(64, activation='tanh', return_sequences=True, return_state=True, dropout=0.2)
all_state_h, state_h, state_c = lstm_encoder(input_layer)
states=[state_h, state_c]
lstm_encoder = layers.LSTM(64, activation='tanh', return_sequences=True, return_state=True, dropout=0.2)
all_state_h, state_h, state_c = lstm_encoder(all_state_h, initial_state=states)
states=[state_h, state_c]
lstm_encoder = layers.LSTM(64, activation='tanh', dropout=0.2)(all_state_h, initial_state=states)
latent = layers.Dense(2, activation='sigmoid', name="Latent_Features")(lstm_encoder)
reshaped_latent = layers.Reshape((1,2))(latent)
reshaped_latent = layers.Concatenate(axis=2)([reshaped_latent, abs_0])
encoder = tf.keras.models.Model(inputs=input_layer, outputs=reshaped_latent, name="Encoder")
encoder.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=1e-3), loss='mae', metrics=['mae'])
encoder.summary()
```

## Подготовка датасета

```python
class DataGenerator:
    def __init__(self, chunk):
        self.chunk = chunk
    def __iter__(self):
        Y = tf.stack([tf.convert_to_tensor(np.log(self.chunk.iloc[:, 0].values * (self.chunk.iloc[:, 1].values + 10000))/25, dtype=tf.float32),
                      tf.convert_to_tensor(np.log(self.chunk.iloc[:, 0].values * (self.chunk.iloc[:, 2].values + 10000))/25, dtype=tf.float32),
                      tf.convert_to_tensor(np.log(self.chunk.iloc[:, 3].values+1)/10, dtype=tf.float32)
        ])
        Y = tf.transpose(Y)
        Y = tf.expand_dims(Y, axis=1)
        abs = []
        phases = []
        for i in range(3, self.chunk.shape[1], 3):
           abs.append(tf.convert_to_tensor(np.log(self.chunk.iloc[:, i].values+1)/10, dtype=tf.float32))
           phases.append(tf.convert_to_tensor((self.chunk.iloc[:, i + 1].values+np.pi)/2/np.pi, dtype=tf.float32))
        abs_tensor = tf.stack(abs, axis=1)
        phases_tensor = tf.stack(phases, axis=1)
        X = tf.stack([abs_tensor, phases_tensor], axis=-1)
        for x, y in zip(X, Y):
          yield (x, y)
```

```python
def filter_dataset(x, y):
  if isinstance(y, tuple):
    return tf.reduce_all(tf.math.is_finite(x)) & tf.reduce_all(tf.math.is_finite(y[0])) & tf.reduce_all(tf.math.is_finite(y[1]))
  else:
    return tf.reduce_all(tf.math.is_finite(x)) & tf.reduce_all(tf.math.is_finite(y))
```

```python
def create_tf_dataset(chunk):
    data_gen = DataGenerator(chunk)
    dataset = tf.data.Dataset.from_generator(
      lambda: data_gen,
        output_signature=(
          tf.TensorSpec(shape=(100, 2), dtype=tf.float32),
          tf.TensorSpec(shape=(1,3), dtype=tf.float32)
        )
    )
    return dataset
```

```python
point_input = layers.Input(shape=(1,4), name="point_input")
lstm_point = layers.LSTM(64, activation='tanh', name='lstm_point', return_sequences=True, return_state=True)
all_state_h, state_h, state_c = lstm_point(point_input)
attention = layers.MultiHeadAttention(num_heads=4, key_dim=64)(all_state_h, all_state_h)
all_state_h = layers.Add()([all_state_h, attention])
states=[state_h, state_c]
lstm_point = layers.LSTM(64, activation='tanh', name='lstm_point2', return_sequences=True, return_state=True)
all_state_h, state_h, state_c = lstm_point(all_state_h, initial_state=states)
attention = layers.MultiHeadAttention(num_heads=4, key_dim=64)(all_state_h, all_state_h)
all_state_h = layers.Add()([all_state_h, attention])
states=[state_h, state_c]
lstm_point = layers.LSTM(64, activation='tanh', name='lstm_point_2')(all_state_h, initial_state=states)
point_abs = layers.Dense(1, activation='sigmoid', name='point_abs')(lstm_point)
point_phase = layers.Dense(1, activation='sigmoid', name='point_phase')(lstm_point)
point_decoder = tf.keras.models.Model(inputs=point_input, outputs={"point_abs":point_abs, "point_phase": point_phase}, name="Point_decoder")
point_decoder.summary()
```

```python
class DataGenerator:
    def __init__(self, chunk):
        self.chunk = chunk
    def __iter__(self):
        X=[]
        Y_abs=[]
        Y_phase=[]
        for j in range(self.chunk.shape[0]):
            for i in range(3, self.chunk.shape[1]-3, 3):
                x=tf.reshape(tf.stack([
                    tf.convert_to_tensor(np.log(self.chunk.iloc[j, 0] * (self.chunk.iloc[j, 1] + 10000)) / 25, dtype=tf.float32),
                    tf.convert_to_tensor(np.log(self.chunk.iloc[j, 0] * (self.chunk.iloc[j, 2] + 10000)) / 25, dtype=tf.float32),
                    tf.convert_to_tensor(np.log(self.chunk.iloc[j, i] + 1) / 10, dtype=tf.float32),
                    tf.convert_to_tensor((self.chunk.iloc[j, i+1] + np.pi) / (2*np.pi), dtype=tf.float32)]), (1,4))
                y_abs=tf.stack([
                    tf.convert_to_tensor(np.log(self.chunk.iloc[j, i+3] + 1) / 10, dtype=tf.float32),
                    tf.convert_to_tensor(self.chunk.iloc[j, i+5], dtype=tf.float32)
                ])
                y_phase = tf.stack([
                    tf.convert_to_tensor((self.chunk.iloc[j, i+4] + np.pi) / 2 / np.pi, dtype=tf.float32),
                    tf.convert_to_tensor(self.chunk.iloc[j, i+5], dtype=tf.float32)
                ])
                X.append(x)
                Y_abs.append(tf.expand_dims(y_abs, axis=0))
                Y_phase.append(tf.expand_dims(y_phase, axis=0))
        for x, y_abs, y_phase in zip(X, Y_abs, Y_phase):
            yield (x, (y_abs, y_phase))
```

## Создание автоэнкодера

```python
class Autoencoder(tf.keras.models.Model):
    def __init__(self, encoder_model, decoder_model):
        super(Autoencoder, self).__init__()
        self.encoder = encoder_model
        self.decoder = decoder_model
    def call(self, inputs):
        params = self.encoder(inputs)
        points = [(params[0][0][2], 0.5)]
        for i in range(99):
            p_x = tf.convert_to_tensor([[params[0][0][0], params[0][0][1], points[-1][0], points[-1][1]]])
            p_x = tf.expand_dims(p_x, axis=1)
            next_point = self.decoder(p_x)
            points.append((tf.squeeze(next_point['point_abs'][0]), tf.squeeze(next_point['point_phase'][0])))
        points_array = tf.convert_to_tensor(points)
        return points_array
    def evaluate(self, dataset, **kwargs):
        total_abs_qty = 0.0
        total_phase_qty = 0.0
        total_count = 0.0
        for x, y in dataset:
          for i in range(x.shape[0]):
              preds = self.call(tf.convert_to_tensor(x[i:i+1, :, :]))
              total_abs_qty += self.abs_qty(tf.convert_to_tensor(y[0]), preds[:, 0])
              total_phase_qty += self.phase_qty(tf.convert_to_tensor(y[1]), preds[:, 1])
              total_count+=1.0
        return total_abs_qty/total_count, total_phase_qty/total_count
    def abs_qty(self, y_true, y_pred):
        qty=y_true[:, :,1]
        mae=tf.abs(y_true[:,:, 0]-y_pred)
        loss=mae*qty
        return tf.reduce_mean(loss)
    def phase_qty(self, y_true, y_pred):
        qty=y_true[:,:,1]
        mae=tf.abs(y_true[:,:, 0]-y_pred)
        loss=mae*qty
        return tf.reduce_mean(loss)
```

## Подготовка датасета

```python
def create_tf_dataset(chunk, line_num=-1):
    data_gen = DataGenerator(chunk, line_num)
    dataset = tf.data.Dataset.from_generator(
      lambda: data_gen,
        output_signature=(
            tf.TensorSpec(shape=(None, 4), dtype=tf.float32),
            (tf.TensorSpec(shape=(1, 2), dtype=tf.float32),
            tf.TensorSpec(shape=(1, 2), dtype=tf.float32))
        )
    )
    dataset=dataset.filter(filter_dataset)
    return dataset
```

```python
class DataGenerator:
    def __init__(self, chunk, len_num):
      self.chunk = chunk
      self.len_num = len_num
    def __iter__(self):
      X=[]
      Y_abs=[]
      Y_phase=[]
      for j in range(self.chunk.shape[0]):
         x1=tf.reshape(tf.convert_to_tensor(np.log(self.chunk.iloc[j, 0] * (self.chunk.iloc[j, 1] + 10000)) / 25, dtype=tf.float32), (1,1))
         x2=tf.reshape(tf.convert_to_tensor(np.log(self.chunk.iloc[j, 0] * (self.chunk.iloc[j, 2] + 10000)) / 25, dtype=tf.float32), (1,1))
         x3=tf.convert_to_tensor(np.log(self.chunk.iloc[j, 3:3*self.len_num+4:3] + 1) / 10, dtype=tf.float32)
         x4=tf.convert_to_tensor((self.chunk.iloc[j, 4:3*self.len_num+5:3] + np.pi) / (2 * np.pi), dtype=tf.float32)
         x4=tf.keras.ops.reshape(x4, (-1, 1))
         x3=tf.keras.ops.reshape(x3, (-1, 1))
         repeat_count = tf.shape(x3)[0]
         x1_repeated = tf.keras.ops.repeat(x1, repeat_count, axis=0)
         x2_repeated = tf.keras.ops.repeat(x2, repeat_count, axis=0)
         x = tf.concat([x1_repeated, x2_repeated, x3, x4], axis=1)
         X.append(x)
         y_abs = tf.stack([tf.convert_to_tensor(np.log(self.chunk.iloc[j, 3*self.len_num + 6] + 1) / 10, dtype=tf.float32),
                           tf.convert_to_tensor(self.chunk.iloc[j, 3*self.len_num + 8], dtype=tf.float32)
         ])
         y_phase = tf.stack([tf.convert_to_tensor((self.chunk.iloc[j, 3*self.len_num + 7] + np.pi) / (2 * np.pi), dtype=tf.float32),
                             tf.convert_to_tensor(self.chunk.iloc[j, 3*self.len_num + 8], dtype=tf.float32)
         ])
         Y_abs.append(tf.expand_dims(y_abs, axis=0))
         Y_phase.append(tf.expand_dims(y_phase, axis=0))
         for x, y_abs, y_phase in zip(X, Y_abs, Y_phase):
             yield (x, (y_abs, y_phase))
```

## Создание автоэнкодера

```python
class Autoencoder(tf.keras.models.Model):
    def __init__(self, encoder, decoder):
        super(Autoencoder, self).__init__()
        self.encoder = encoder
        self.decoder = decoder
    def call(self, inputs):
        params = self.encoder(inputs)
        points = [(params[0][0][2], 0.5)]
        for i in range(99):
            p_x=[]
            for u in points:
                p_x.append([params[0][0][0], params[0][0][1], u[0], u[1]])
            p_x=tf.convert_to_tensor(p_x)
            p_x=tf.expand_dims(p_x, axis=0)
            next_point=self.decoder(p_x)
            points.append((tf.squeeze(next_point['point_abs']), tf.squeeze(next_point['point_phase'])))
        points_array = tf.convert_to_tensor(points)
        return points_array
    def evaluate(self, dataset, **kwargs):
        total_abs_qty = 0.0
        total_phase_qty = 0.0
        total_count = 0.0
        for x, y in dataset:
          for i in range(x.shape[0]):
              preds = self.call(tf.convert_to_tensor(x[i:i+1, :, :]))
              total_abs_qty += self.abs_qty(tf.convert_to_tensor(y[0]), preds[:, 0])
              total_phase_qty += self.phase_qty(tf.convert_to_tensor(y[1]), preds[:, 1])
              total_count+=1.0
        return total_abs_qty/total_count, total_phase_qty/total_count
    def abs_qty(self, y_true, y_pred):
        qty=y_true[:, :,1]
        mae=tf.abs(y_true[:,:, 0]-y_pred)
        loss=mae*qty
        return tf.reduce_mean(loss)
    def phase_qty(self, y_true, y_pred):
        qty=y_true[:,:,1]
        mae=tf.abs(y_true[:,:, 0]-y_pred)
        loss=mae*qty
        return tf.reduce_mean(loss)
```
